## ガバナンスの基本

**この章で覚えること**

- **Organizations**
- **SCP**
- **Control Tower**

### AWS Organizations

**AWS Organizations**

複数のAWSアカウントを一元管理するためのサービスです。

- **OU**：アカウントの階層管理
- **SCP**：組織全体の最大権限（許可の上限）
- **コスト配分タグ**：部門・プロジェクト別のコスト把握

### AWS Control Tower

**AWS Control Tower**

マルチアカウント環境を**ランディングゾーン**（ベストプラクティスに沿った初期構成）でセットアップし、ガバナンスを自動化するサービスです。

新規アカウント作成時にセキュリティ・ログ設定を一貫して適用できます。

## 管理とモニタリング

**この章で覚えること**

- **CloudWatch**
- **Systems Manager**
- **EventBridge**

### Amazon CloudWatch

**Amazon CloudWatch**

AWSリソースとアプリケーションのモニタリングを行うサービスです。

メトリクスの収集、ログの監視、アラームの設定が可能で、システムのパフォーマンスや稼働状態をリアルタイムで把握できます。

### AWS Systems Manager

**AWS Systems Manager**

運用の自動化と管理を支援する統合ツールです。

特に、複数のAWSアカウントやリージョンにまたがる環境での一貫した管理が実現できます。

### AWS Trusted Advisor

**AWS Trusted Advisor**

AWS環境の最適化を支援するサービスです。

- コスト・セキュリティ・性能などの観点で改善提案
- 未使用 EC2、古い S3 バケットなどの指摘

### AWS Compute Optimizer

**AWS Compute Optimizer**

コンピューティングリソースの使用状況を分析し、最適なリソース構成を提案します。

### Amazon Managed Grafana と Amazon Managed Service for Prometheus

**Amazon Managed Grafana**と**Amazon Managed Service for Prometheus**は、メトリクスを可視化・収集するマネージドの可観測性ツールです。

CloudWatchと併用し、複雑なダッシュボードやPrometheus互換の監視に向きます。

### Amazon EventBridge

**Amazon EventBridge**

AWSサービスやSaaSからのイベントをルールで受け取り、Lambda・SNS等へ**イベント駆動**で連携するサービスです（旧CloudWatch Eventsの後継）。

### 操作インターフェース（補足）

- **AWS Management Console**：ブラウザでリソースを操作（初学者向け）
- **AWS CLI**：コマンドやスクリプトで自動化

## セキュリティとコンプライアンス

**この章で覚えること**

- **Config** vs **CloudTrail** の役割の違い

### AWS Config

**AWS Config**

AWSリソースの設定変更を追跡し、コンプライアンスを維持するサービスです。

Configルールを設定することで、リソースの設定がベストプラクティスに従っているかを自動的に評価し、違反が検出された場合にはアラートを発生させます。

### AWS CloudTrail

**AWS CloudTrail**

AWSアカウント内の**APIコール**を記録し、監査やセキュリティ分析のための履歴を提供します。

**AWS Config と CloudTrail（比較）**

| サービス | 記録するもの |
|----------|--------------|
| **CloudTrail** | 誰が・いつ・どのAPIを実行したか |
| **Config** | リソース設定の変更とコンプライアンスルール評価 |

**補足（優先度低）**：**AWS License Manager**（ライセンス管理）、**AWS Proton**（テンプレートによるデプロイ自動化）はSAAでは概要把握で十分です。

## コスト管理と最適化

**この章で覚えること**

- **Cost Explorer**
- **Budgets**
- タグ

### AWS Cost Explorer

**AWS Cost Explorer**

コストと使用状況をグラフ化・分析するツールです。

タグやサービス別の内訳把握に使います。

### AWS Budgets

**AWS Budgets**

コスト・使用量・リザーブド利用率などに**予算とアラート**を設定するサービスです（Cost Explorerが「分析」、Budgetsが「予算超過の通知」）。

### AWS Well-Architected Tool

**AWS Well-Architected Tool**

アーキテクチャを6つの柱（運用上の優秀性、セキュリティ、信頼性、パフォーマンス効率、コスト最適化、持続可能性）でレビューします。

**Savings Plans**（コンピューティング章と関連）：1〜3年の利用コミットでオンデマンドより割引。Compute Savings Plansはファミリー横断で適用可能です。

### AWS Service Catalog

**AWS Service Catalog**

標準化された製品をカタログ化し、迅速かつ一貫したデプロイを可能にします。

> ### 試験で押さえる設計の考え方（本章関連）
>
>

> - **統制**：Organizations + SCP + Control Tower
> - **可視性**：CloudWatch + CloudTrail + Config
> - **コスト**：Cost Explorer + Budgets + タグ

## まとめ

- **Organizations / Control Tower**：マルチアカウント統制。OU・SCP・ランディングゾーン
- **EventBridge**：イベント駆動の運用自動化
- **CloudWatch / Systems Manager / Trusted Advisor / Compute Optimizer**：監視と改善提案
- **Config / CloudTrail**：設定の追跡とAPI監査
- **Cost Explorer / Budgets**：コスト分析と予算アラート
- **Well-Architected Tool**：設計レビューの枠組み

## 復習クイズ　解答・解説

マルチアカウント統合、請求統合、OU管理。→ ガバナンスの基本

**A2.** **SCP（Service Control Policy）**

**A3.** **AWS CloudTrail**  
誰がいつどのAPIを呼んだかの監査ログ。→ セキュリティとコンプライアンス

設定変更の記録とコンプライアンスルール評価。→ セキュリティとコンプライアンス

**A5.** **AWS Cost Explorer**（分析） / **AWS Budgets**（予算・アラート）  
Cost Explorerで傾向把握、Budgetsで超過通知。→ コスト管理と最適化

**A6.** **Amazon CloudWatch**  
メトリクス・ログ・アラーム。→ 管理とモニタリング

**A7.** **Amazon EventBridge**  
イベント駆動の連携。CloudWatch Eventsの後継。→ Amazon EventBridge
