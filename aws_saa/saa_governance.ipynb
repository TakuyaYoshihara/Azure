## ガバナンスの基本
ガバナンスの基本は、クラウドリソースを効率的に管理し、統制することです。ここでは、**AWS Organizations**と**AWS Control Tower**を中心に説明します。

### AWS Organizations


**AWS Organizations**は、複数のAWSアカウントを一元管理するためのサービスです。部門別アカウント、請求の統合、**OU（Organizational Unit）**による階層管理、**SCP（Service Control Policy）**による最大権限の制限（例：特定リージョンのみ許可）が可能です。**リソースタグ**と**コスト配分タグ**を組み合わせると、部門別コスト把握がしやすくなります。

### AWS Control Tower


**AWS Control Tower**は、マルチアカウント環境を**ランディングゾーン**（ベストプラクティスに沿った初期構成）でセットアップし、ガバナンスを自動化するサービスです。新規アカウント作成時にセキュリティ・ログ設定を一貫して適用できます。

## 管理とモニタリング
管理とモニタリングは、リソースの状態を常に把握し、問題が発生した際に迅速に対応するために重要です。これにより、運用の可視性が向上し、効率的なリソース管理が可能になります。

### Amazon CloudWatch


**Amazon CloudWatch**は、AWSリソースとアプリケーションのモニタリングを行うサービスです。メトリクスの収集、ログの監視、アラームの設定が可能で、システムのパフォーマンスや稼働状態をリアルタイムで把握できます。CloudWatchを使用することで、リソースの過不足を早期に検知し、必要な対策を講じることができます。これにより、リソースの最適な配分が可能となり、無駄なコストを削減できます。

### AWS Systems Manager


**AWS Systems Manager**は、運用の自動化と管理を支援する統合ツールです。これにより、パッチ管理、インベントリ収集、オートメーションの実行が容易に行えます。特に、複数のAWSアカウントやリージョンにまたがる環境での一貫した管理が実現できます。Systems Managerを活用することで、日常的な運用タスクを自動化し、運用効率を大幅に向上させることが可能です。これにより、運用の一貫性が保たれ、手動でのエラーを減少させることができます。

### AWS Trusted Advisor


**AWS Trusted Advisor**は、AWS環境の最適化を支援するサービスです。セキュリティ、コスト、パフォーマンスのベストプラクティスに基づき、リソースの最適化提案を提供します。例えば、未使用のEC2インスタンスの特定や古いS3バケットの削除を提案し、コスト削減に寄与します。Trusted Advisorの提案を実行することで、運用コストの削減とセキュリティの強化が可能です。これにより、企業は持続的に最適なリソース利用を実現できます。

### AWS Compute Optimizer


**AWS Compute Optimizer**は、コンピューティングリソースの使用状況を分析し、最適なリソース構成を提案します。これにより、オーバープロビジョニングの回避とコスト削減が実現します。

### Amazon Managed Grafana と Amazon Managed Service for Prometheus


**Amazon Managed Grafana**と**Amazon Managed Service for Prometheus**は、メトリクスを可視化・収集するマネージドの可観測性ツールです。CloudWatchと併用し、複雑なダッシュボードやPrometheus互換の監視に向きます。

### Amazon EventBridge


**Amazon EventBridge**は、AWSサービスやSaaSからのイベントをルールで受け取り、Lambda・SNS等へ**イベント駆動**で連携するサービスです（旧CloudWatch Eventsの後継）。

### 操作インターフェース（補足）
- **AWS Management Console**：ブラウザでリソースを操作（初学者向け）
- **AWS CLI**：コマンドやスクリプトで自動化

## セキュリティとコンプライアンス
セキュリティとコンプライアンスは、ガバナンスの要です。AWSのツールを活用することで、企業はセキュリティの強化とコンプライアンスの維持を両立できます。

### AWS Config


**AWS Config**は、AWSリソースの設定変更を追跡し、コンプライアンスを維持するサービスです。Configルールを設定することで、リソースの設定がベストプラクティスに従っているかを自動的に評価し、違反が検出された場合にはアラートを発生させます。これにより、継続的なコンプライアンスの維持が可能です。

### AWS CloudTrail


**AWS CloudTrail**は、AWSアカウント内の**APIコール**を記録し、監査やセキュリティ分析のための履歴を提供します。

**AWS Config と CloudTrail（比較）**

| サービス | 記録するもの |
|----------|--------------|
| **CloudTrail** | 誰が・いつ・どのAPIを実行したか |
| **Config** | リソース設定の変更とコンプライアンスルール評価 |

> **補足（優先度低）**：**AWS License Manager**（ライセンス管理）、**AWS Proton**（テンプレートによるデプロイ自動化）はSAAでは概要把握で十分です。

## コスト管理と最適化
コスト管理と最適化は、AWS環境を効率的に運用するために重要です。適切なツールを使用することで、コストを最小限に抑えつつ、パフォーマンスを最大化できます。

### AWS Cost Explorer


**AWS Cost Explorer**は、コストと使用状況をグラフ化・分析するツールです。タグやサービス別の内訳把握に使います。

### AWS Budgets


**AWS Budgets**は、コスト・使用量・リザーブド利用率などに**予算とアラート**を設定するサービスです（Cost Explorerが「分析」、Budgetsが「予算超過の通知」）。

### AWS Well-Architected Tool


**AWS Well-Architected Tool**は、アーキテクチャを6つの柱（運用上の優秀性、セキュリティ、信頼性、パフォーマンス効率、コスト最適化、持続可能性）でレビューします。

**Savings Plans**（コンピューティング章と関連）：1〜3年の利用コミットでオンデマンドより割引。Compute Savings Plansはファミリー横断で適用可能です。

### AWS Service Catalog


**AWS Service Catalog**は、標準化された製品をカタログ化し、迅速かつ一貫したデプロイを可能にします。これにより、コスト管理とコンプライアンスが確保され、組織全体での一貫したリソース利用が可能です。Service Catalogを使用することで、組織全体での効率的なリソース利用と標準化が推進されます。

### 試験で押さえる設計の考え方（本章関連）
- **統制**：Organizations + SCP + Control Tower
- **可視性**：CloudWatch + CloudTrail + Config
- **コスト**：Cost Explorer + Budgets + タグ

## まとめ
- **Organizations / Control Tower**：マルチアカウント統制。OU・SCP・ランディングゾーン
- **EventBridge**：イベント駆動の運用自動化
- **CloudWatch / Systems Manager / Trusted Advisor / Compute Optimizer**：監視と改善提案
- **Config / CloudTrail**：設定の追跡とAPI監査
- **Cost Explorer / Budgets**：コスト分析と予算アラート
- **Well-Architected Tool**：設計レビューの枠組み
