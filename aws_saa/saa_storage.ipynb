# ストレージ
本章では、AWSソリューションアーキテクトアソシエイト試験に必要なストレージサービスについて学習します。

## 本章の構成
- ブロックストレージ
- ファイルストレージ
- オブジェクトストレージ
- ハイブリッドストレージ
- 高性能ファイルシステム

## ブロックストレージ
**Amazon Elastic Block Store (Amazon EBS)** は、EC2インスタンスにアタッチして使用する高性能で低レイテンシのブロックストレージサービスです。EBSは、データの持続性を確保し、汎用SSD（**gp2/gp3**）、プロビジョンドIOPS SSD（**io1/io2**）、スループット最適化HDD（**st1**）、Cold HDD（**sc1**）の4つのボリュームタイプを提供します。

EBSの主なユースケースには、データベースのストレージや高パフォーマンスが求められるアプリケーションのバックエンドが含まれます。例えば、**io1/io2**は、トランザクションが頻繁に発生するデータベースに適しています。一方、**st1**や**sc1**は、ログデータやバックアップデータなど、大容量のデータ保存に適しています。

EBSの重要な機能として、**スナップショット**があります。これは、データのバックアップを取得し、別のリージョンに複製することが可能で、災害復旧に役立ちます。また、EBSは**データの暗号化**をサポートしており、データのセキュリティを強化します。

## ファイルストレージ
**Amazon Elastic File System (Amazon EFS)** は、完全マネージド型のNFSファイルシステムを提供するサービスです。EFSは、複数のEC2インスタンスから同時にアクセスでき、スケーラブルで高い可用性を持つストレージです。

EFSの主なユースケースには、コンテンツ管理システムや、ホームディレクトリの共有ストレージが含まれます。例えば、ウェブサーバーが複数台ある場合、これらが同じデータにアクセスする必要がある時にEFSは非常に有効です。EFSは**自動的に容量を拡張または縮小**するため、使用量に応じてストレージを最適化できます。

EFSには、**標準ストレージクラス**と**インフリークエントアクセスストレージクラス**があり、頻繁にアクセスされるデータとまれにアクセスされるデータを分けることで、コストを最適化できます。

## オブジェクトストレージ
**Amazon S3** は、スケーラブルで高可用性を持つオブジェクトストレージサービスです。S3は、データの保存、取得、アーカイブに適しており、バケット単位でデータを管理します。S3は、**バージョニング**、**ライフサイクルポリシー**、**アクセスコントロールリスト（ACL）**、**バケットポリシー**などの機能を提供し、データの管理を効率化します。

S3の主なユースケースには、ウェブサイトのホスティング、データバックアップ、ビッグデータ分析のデータレイクが含まれます。例えば、S3は静的ウェブサイトのホスティングに利用されることが多く、コンテンツをユーザーに高速で配信できます。また、**バージョニング**を利用することで、データの変更履歴を管理し、誤って削除したデータの復元が可能です。

**Amazon S3 Glacier** は、長期間アクセスしないデータのアーカイブに特化した低コストのストレージサービスです。S3 Glacierは、バックアップデータや法令遵守のために長期間保存する必要があるデータに適しています。データの取得には数時間かかる場合がありますが、非常に安価な料金でデータを長期間保存できます。

## ハイブリッドストレージ
**AWS Storage Gateway** は、オンプレミスのストレージ環境とAWSクラウドストレージを統合するサービスです。これにより、オンプレミスのアプリケーションからクラウドストレージに直接アクセスできます。Storage Gatewayは、ファイルゲートウェイ、ボリュームゲートウェイ、テープゲートウェイの3つのモードを提供し、さまざまなストレージニーズに対応します。

Storage Gatewayのユースケースには、オンプレミスのデータのバックアップやアーカイブ、クラウドへのデータ移行が含まれます。例えば、企業がオンプレミスのファイルサーバーをAWSにバックアップする際に、Storage Gatewayを利用することで、シームレスにクラウドと連携できます。

## 高性能ファイルシステム
**Amazon FSx** は、高性能なファイルシステムを提供するサービスです。FSxには、Windowsファイルサーバー向けの**Amazon FSx for Windows File Server**、高性能なLustreファイルシステム向けの**Amazon FSx for Lustre**、NetApp ONTAP向けの**Amazon FSx for NetApp ONTAP**、OpenZFS向けの**Amazon FSx for OpenZFS**があります。

例えば、FSx for Windows File Serverは、Windowsベースのアプリケーションに対して高い互換性を提供し、Active Directoryと統合可能です。一方、FSx for Lustreは、高性能コンピューティング（HPC）やビッグデータ解析のワークロードに適しています。FSx for NetApp ONTAPは、NetAppの機能をクラウドで利用するために最適です。

## まとめ
AWSのストレージサービスは、多様なニーズに対応するために設計されています。**Amazon EBS**は、高性能で低レイテンシのブロックストレージを提供し、データのバックアップと暗号化が可能です。**Amazon EFS**は、複数のEC2インスタンスから同時にアクセスできるスケーラブルなファイルストレージを提供します。**Amazon S3**は、オブジェクトストレージとして高可用性とスケーラビリティを提供し、**S3 Glacier**は低コストで長期間のデータアーカイブに適しています。**AWS Storage Gateway**は、オンプレミスとクラウドの統合をサポートし、**Amazon FSx**は特定のワークロードに最適な高性能ファイルシステムを提供します。
