# ネットワーク

本章では、AWSソリューションアーキテクトアソシエイト試験に必要なネットワークサービスについて学習します。

## 本章の構成
- Elastic Load Balancing (ELB)
- AWS Global Accelerator
- AWS PrivateLink

## Elastic Load Balancing (ELB)
Elastic Load Balancing (ELB) は、AWS上でのトラフィック分散サービスです。ELBは、複数のEC2インスタンスやその他のターゲットにトラフィックを自動的に分散し、高可用性と耐障害性を確保します。

ELB についてご説明する前に、**OSI基本参照モデル**について振り返っておきましょう。OSI基本参照モデルは、通信体系を7つの階層に分け、標準化したモデルでした。各層が独立して機能し他の層と組み合わせることで通信プロセスが成立します。層ごとにプロトコルを分けることでプロトコルの役割が明確になり、エラーの特定や変更があっても最小限のプロトコル変更で対応できます。

![](../../../images/internal/platform/course/cloud-aws/osi-sansho-model.png)

ELB には、以下の3つの主要なタイプがあり、それぞれ動作する層が異なっています。

- **Application Load Balancer (ALB)**: 第7層（アプリケーション層）で動作し、HTTPおよびHTTPSトラフィックのロードバランシングを行います。ALBは、ターゲットグループごとにトラフィックを分配し、各ターゲットのヘルスチェックを実行します。URLベースのルーティング、WebSocketのサポート、シングルページアプリケーション（SPA）のサポートなどの機能があります。
- **Network Load Balancer (NLB)**: 第4層（トランスポート層）で動作し、TCPおよびUDPトラフィックのロードバランシングを行います。NLBは、高いスループットと低レイテンシを提供し、ミリ秒単位のヘルスチェックと高スケーラビリティを実現します。IPアドレスベースのルーティングや静的IPのサポートも特徴です。
- **Classic Load Balancer (CLB)**: 第4層および第7層で動作し、EC2-Classicネットワークでの使用を推奨される旧式のロードバランサーです。ALBやNLBに比べて機能が限定されていますが、依然として多くのレガシーアプリケーションで利用されています。

ELBの重要な機能には、ヘルスチェック、SSL/TLS終端、コンテナサービス（Amazon ECS、Amazon EKS）との統合、Auto Scalingとの連携があります。これにより、アプリケーションの高可用性とスケーラビリティを実現できます。重要なポイントとして、ELBの3種類（ALB、NLB、CLB）の特徴と用途を理解し、ヘルスチェックとSSL/TLS終端の役割を把握することが求められます。また、Auto Scalingとの連携による高可用性の実現方法を学ぶことも重要です。

- **ヘルスチェック**: 各ターゲットの正常性を監視し、不具合があればトラフィックを振り分けない機能
- **SSL/TLS終端**: 安全な通信のための暗号化と復号化の処理を行う機能

## AWS Global Accelerator
AWS Global Acceleratorは、インターネットトラフィックを最適化し、ユーザーに対して低遅延で高可用性の接続を提供するサービスです。AWSのグローバルネットワークを利用し、エッジロケーションを経由してトラフィックを最適なエンドポイントにルーティングします。

主な特徴としては、複数のリージョンにまたがるエンドポイントへのトラフィック分散、トラフィックの自動再ルーティング、静的IPアドレスの提供、エンドポイントグループごとのトラフィック分配の設定が可能です。これにより、アプリケーションのパフォーマンスと可用性が向上します。例えば、ユーザーの接続が最も近いエッジロケーションを通じて行われるため、遅延が大幅に低減されます。また、障害が発生した場合でも、トラフィックが自動的に別の健全なエンドポイントにルーティングされるため、サービスの可用性が確保されます。

AWS Global Acceleratorを使用する際の重要なポイントとしては、トラフィックの分散方法と再ルーティングの仕組みを理解すること、静的IPアドレスの提供による利便性を把握することが挙げられます。これらにより、グローバルなユーザーに対して安定した接続を提供することが可能です。

- **エッジロケーション**: トラフィックの最適化を行うためのAWSのデータセンター
- **静的IPアドレス**: 変わらない固定のIPアドレス

## AWS PrivateLink
AWS PrivateLinkは、VPC内でAWSサービスやオンプレミスのアプリケーションへのプライベート接続を提供するサービスです。インターネットを経由せずに、安全にデータを送受信することができます。PrivateLinkを使用することで、VPCエンドポイントを介して、プライベートIPアドレスを使用してサービスにアクセスすることができます。

主な利点には、セキュリティの向上、ネットワークレイテンシの低減、簡単な設定と管理があります。PrivateLinkは、企業が自社のVPC内でAWSサービスやサードパーティサービスにアクセスする際に、セキュリティとパフォーマンスを向上させる手段として広く利用されています。例えば、VPC外の通信が不要になり、データの漏洩リスクが低減されます。また、インターネットを経由しないため、ネットワークレイテンシが低くなります。設定と管理も、AWS Management Console、CLI、SDKを使用して簡単に行えます。

AWS PrivateLinkを利用する際の重要なポイントは、セキュリティ向上のメリットとネットワークレイテンシの低減効果を理解することです。また、設定と管理が簡単である点も押さえておくべきです。

- **VPC（Virtual Private Cloud）**: 仮想プライベートクラウド
- **レイテンシ**: データの送受信にかかる時間

## まとめ
AWSのネットワークサービスは、トラフィックの効率的な分散、最適化された接続、高度なセキュリティを提供します。Elastic Load Balancing (ELB)は高可用性とスケーラビリティを実現し、AWS Global Acceleratorは低遅延と高可用性の接続を提供し、AWS PrivateLinkはセキュアなプライベート接続を提供します。これらのサービスの特徴を理解し、適切に活用することで、AWS環境での効率的なネットワーク設計と運用が可能になります。特に、それぞれのサービスの特徴、用途、利点をしっかりと把握し、具体的なシナリオに応じて適切に選択・運用することが重要です。
