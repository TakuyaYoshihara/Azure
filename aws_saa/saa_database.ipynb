## リレーショナルデータベース
リレーショナルデータベースは、構造化データを格納し、SQLを用いてデータを管理するシステムです。AWSは、幅広いリレーショナルデータベースサービスを提供しています。

### Amazon RDS


**Amazon RDS（Relational Database Service）**は、クラウド上でリレーショナルデータベースを簡単にセットアップ、運用、およびスケーリングするためのサービスです。MySQL、PostgreSQL、MariaDB、Oracle、SQL Serverなどをサポートします。**自動バックアップ**、**手動スナップショット**、メンテナンスウィンドウ、パッチ適用をAWSが代行します。

### Amazon Aurora


**Amazon Aurora**は、MySQLおよびPostgreSQL互換の高性能RDBです。ストレージは自動拡張し、AZ内で6コピー保持されます。**Aurora Global Database**は、プライマリリージョンから最大5つのセカンダリリージョンへレプリケーションし、DRと低RPOを実現します（試験頻出）。

### Amazon Aurora Serverless


**Amazon Aurora Serverless**は、負荷に応じてキャパシティを自動調整するオプションです（現在は **Aurora Serverless v2** が一般的）。断続的なアクセスや予測しづらい負荷向けです。

**Multi-AZ とリードレプリカ（試験頻出）**

| 機能 | 目的 |
|------|------|
| **Multi-AZ** | 障害時の自動フェイルオーバー（可用性） |
| **リードレプリカ** | 読み取り負荷の分散（最大5台など） |

- **向く例（RDS/Aurora）**：トランザクション、既存RDB移行、複雑なSQL
- **向かない例**：秒間数百万のキー・バリューアクセス → DynamoDB

## NoSQLデータベース
NoSQLデータベースは、構造化データ、半構造化データ、および非構造化データを柔軟に管理するためのシステムです。AWSは、スケーラブルで高性能なNoSQLデータベースを提供しています。

### Amazon DynamoDB


**Amazon DynamoDB**は、フルマネージドなNoSQL（キー・バリュー）データベースで、ミリ秒台の応答が特徴です。**オンデマンド**（使った分）と**プロビジョンド**（RCU/WCUを指定）のキャパシティモードがあります。DynamoDB Streamsで変更イベントをLambda等に連携できます。

- **DynamoDB Accelerator（DAX）**：読み取り専用キャッシュで読取レイテンシをさらに短縮
- **グローバルテーブル**：マルチリージョンでレプリケーション（マルチマスター）
- **DocumentDB** との違い：JSONドキュメント・MongoDB互換が必要なら DocumentDB、キー・バリューでスケールなら DynamoDB

### Amazon DocumentDB (MongoDB互換)


**Amazon DocumentDB（MongoDB互換）**は、スケーラブルで高可用性のドキュメントデータベースサービスです。DocumentDBは、MongoDBのAPIと互換性があり、MongoDBデータベースからの移行を容易にします。フルマネージドなサービスであり、バックアップやスケーリング、パッチ適用などの管理タスクを自動化します。

### Amazon Keyspaces (Apache Cassandra向け)


**Amazon Keyspaces（Apache Cassandra向け）**は、スケーラブルで高可用性のNoSQLデータベースサービスで、Apache Cassandra互換のAPIを提供します。Keyspacesは、Cassandraクラスターの管理を簡素化し、インフラストラクチャの運用を軽減します。

## インメモリキャッシュ
インメモリキャッシュは、データのアクセス速度を向上させるために使用される高速キャッシュシステムです。AWSは、スケーラブルなインメモリキャッシュサービスを提供しています。

### Amazon ElastiCache


**Amazon ElastiCache**は、RedisおよびMemcachedのインメモリキャッシュです。

| エンジン | 特徴 | 向く例 |
|----------|------|--------|
| **Redis** | 永続化、レプリケーション、データ構造が豊富 | セッション、ランキング、Pub/Sub |
| **Memcached** | シンプルなキー・バリューキャッシュ | 読み取りキャッシュのみ |

## グラフデータベース
グラフデータベースは、ノードとエッジを用いてデータの関係性を表現するデータベースです。AWSは、グラフデータの複雑なクエリを効率的に処理するサービスを提供しています。

### Amazon Neptune


**Amazon Neptune**は、フルマネージドなグラフデータベースサービスで、グラフモデルに基づいたデータの格納とクエリをサポートします。Neptuneは、Property GraphとRDF（Resource Description Framework）をサポートし、複雑なグラフクエリを効率的に実行します。高可用性と耐障害性を備えており、大規模なグラフデータセットに対しても優れたパフォーマンスを発揮します。

グラフデータベースにおいて特に重要なのは、Neptuneのグラフモデルサポートと高可用性です。これを理解することで、関係性データを効率的に管理し、分析することが可能になります。

## 台帳・監査向けデータベース

### Amazon Quantum Ledger Database (Amazon QLDB)


**Amazon QLDB**は、**改ざん不能な台帳（レジャー）**として全変更履歴を記録するデータベースです。金融取引や監査証跡など、履歴の完全性が重要な用途向けです（一般的なトランザクションDBの代替ではありません）。

## データウェアハウス
データウェアハウスは、大量のデータを分析目的で集約し、複雑なクエリを高速で実行するシステムです。AWSは、スケーラブルで高性能なデータウェアハウスサービスを提供しています。

### Amazon Redshift


**Amazon Redshift**は、OLAP（分析）向けのデータウェアハウスです。RDSなどの**OLTP（取引処理）**とは用途が異なります。Redshift SpectrumでS3上のデータもSQLから参照できます。

- **向く例**：BI、集計レポート、ログ分析
- **向かない例**：1件ずつの更新が多い業務システム → RDS/Aurora

**S3上のデータ分析（試験補足）**

- **Amazon Athena**：S3のデータに対してサーバーレスでSQLクエリ（ログ分析等）
- **AWS Glue**：ETLジョブとデータカタログ（メタデータ管理）

## コスト最適化とキャパシティプランニング
AWSのデータベースソリューションを設計する際には、コスト最適化とキャパシティプランニングが重要です。これにより、効率的かつ費用対効果の高いデータベース運用が可能になります。

### コスト管理サービス
AWSは、コスト管理サービスとして**コスト配分タグ**や**マルチアカウント請求**を提供しています。これらの機能を使用することで、コストを正確に追跡し、部門やプロジェクトごとにコストを配分できます。

### コスト管理ツール
AWSのコスト管理ツールには、**Cost Explorer**、**AWS Budgets**、**AWS Cost and Usage Report**などがあります。Cost Explorerを使用することで、コストの傾向を視覚化し、予算を設定して監視できます。AWS Budgetsを利用することで、予算超過の際にアラートを受け取ることができます。AWS Cost and Usage Reportは、詳細なコストと使用状況のレポートを提供し、分析に役立ちます。

### キャッシュ戦略
インメモリキャッシュを使用することで、データベースの負荷を軽減し、コストを削減できます。**Amazon ElastiCache**は、キャッシュ戦略の実装に役立つサービスで、RedisとMemcachedをサポートしています。

### データ保持ポリシー
適切なデータ保持ポリシーを設計することで、不要なデータを削減し、ストレージコストを最適化できます。データのライフサイクル管理を通じて、古いデータをアーカイブまたは削除することで、コストを管理します。

### キャパシティプランニング
サービスごとに単位が異なります。**DynamoDB**は **RCU/WCU**（またはオンデマンド）、**Aurora Serverless**は **ACU** などで計画します。

### データベース接続とプロキシ
データベース接続とプロキシの最適化も重要です。**Amazon RDS Proxy**は、データベース接続の管理と最適化を支援するサービスで、アプリケーションのスケーラビリティと可用性を向上させます。

### データベースエンジンの適切なユースケース
AWSは、多様なデータベースエンジンを提供しています。例えば、**MySQL**と**PostgreSQL**の比較では、特定のユースケースに適したエンジンを選択することが重要です。MySQLは、読み込み性能に優れ、一方でPostgreSQLは、複雑なクエリやデータの整合性を重視するアプリケーションに向いています。

### 試験で押さえる設計の考え方（本章関連）
- **可用性**：Multi-AZ（RDS）、Aurora Global Database（DR）
- **性能**：リードレプリカ、ElastiCache、DAX
- **分析**：Redshift（OLAP）、Athena（S3 SQL）

## まとめ
- **RDS / Aurora**：SQL・トランザクション。Multi-AZ＝可用性、リードレプリカ＝読取分散
- **Aurora Global Database / DAX**：DR・読取高速化
- **DynamoDB**：スケールするNoSQL。オンデマンド vs プロビジョンド
- **Athena / Glue**：S3分析・ETL（試験補足）
- **ElastiCache**：Redis vs Memcached でキャッシュ戦略
- **Neptune**：グラフ（関係性）データ
- **QLDB**：改ざん不能な台帳・監査
- **Redshift**：分析（OLAP）。RDS（OLTP）と用途を混同しない
- **RDS Proxy**：接続プールでスケール時の接続枯渇を緩和

## 復習クイズ　解答・解説

**A1.** **Multi-AZ**  
スタンバイへの自動フェイルオーバー（可用性）。読取分散ではない。→ リレーショナルデータベース

**A2.** **リードレプリカ（Read Replica）**  
読み取り専用の複製で読取負荷を分散。→ リレーショナルデータベース / コスト最適化節  
※よくある誤答：Multi-AZ（可用性が目的）

**A3.** **Amazon DynamoDB**  
スケーラブルなNoSQL。RDSはSQL・トランザクション向け。→ NoSQLデータベース

**A4.** **Amazon RDS**（または **Amazon Aurora**）  
マネージドRDB。→ リレーショナルデータベース

**A5.** **Redis**  
永続化・レプリケーション・豊富なデータ構造。Memcachedはシンプルなキャッシュのみ。→ インメモリキャッシュ

**A6.** **Amazon Redshift**  
OLAP・分析。RDSはOLTP（取引処理）。→ データウェアハウス

**A7.** **Amazon QLDB（Quantum Ledger Database）**  
改ざん不能な変更履歴・台帳。一般的な分散DBの代替ではない。→ 分散データベース

**A8.** **Amazon RDS Proxy**  
接続プールでスケール時の接続枯渇を緩和。→ コスト最適化とキャパシティプランニング

**A9.** **DynamoDB Accelerator（DAX）**  
DynamoDB専用の読取キャッシュ。→ Amazon DynamoDB（DAXの説明）
