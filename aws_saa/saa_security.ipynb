## アイデンティティとアクセス管理
AWSのアイデンティティとアクセス管理サービスは、ユーザーとリソースのセキュリティを確保するために不可欠です。これらのサービスを使用することで、アクセス制御、ユーザー管理、および認証を効率的に行うことができます。

### AWS Identity and Access Management (IAM)


**AWS Identity and Access Management (IAM)**は、AWSリソースへのアクセスを制御するサービスです。ユーザー、グループ、ロールに**ポリシー（JSONの権限定義）**を割り当て、**最小権限の原則**（必要な操作だけ許可）で設計します。

| 要素 | 役割 | 例 |
|------|------|-----|
| ユーザー | 人やアプリ用の恒久ID | 開発者アカウント |
| グループ | ユーザーをまとめてポリシー付与 | Developers |
| ロール | 一時的な権限。サービスや他アカウント向け | EC2がS3へアクセス |
| ポリシー | 許可/拒否のルール | `s3:GetObject` のみ許可 |

**試験の定番**：EC2やLambdaからS3へアクセスするときは、アクセスキーを埋め込まず**IAMロール**を使います。

**IAMポリシー vs リソースベースポリシー**：IAMポリシーはユーザー/ロールに付与。S3の**バケットポリシー**のようにリソース側に付与するものを**リソースベースポリシー**という。両方が評価される。

**Permission Boundary**：ユーザー/ロールに「最大限これ以上の権限は付与できない」上限を設定する。ルートユーザーは日常運用では使わず、MFA必須とする。

IAMユーザーとルートユーザーには、**多要素認証（MFA）**を設定することで、セキュリティを強化します。MFAを使用することで、認証プロセスに追加のセキュリティレイヤーを追加し、不正アクセスのリスクを減らします。

IAMでは、ユーザー、グループ、ロール、ポリシーを組み合わせることで、柔軟な認証モデルを設計できます。**IAMロール**は、一時的なアクセス権を提供するため、特定のタスクを実行する際に非常に有用です。ロールは、**AWS Security Token Service（AWS STS）**を使用して一時的なセキュリティ資格情報を取得できます。

IAMロールは、異なるAWSアカウント間でアクセスを共有するために使用されます。**クロスアカウントアクセス**を設定することで、一つのアカウントのリソースに対して他のアカウントのユーザーがアクセスできるようになります。また、**ロールスイッチング**を使用することで、ユーザーは一時的に別のロールを引き受けて、そのロールに関連する権限を使用できます。

### AWS IAM Identity Center (AWS Single Sign-On)


**AWS IAM Identity Center（AWS Single Sign-On）**は、複数のAWSアカウントやサードパーティアプリケーションへの**シングルサインオン（SSO）**を提供します。ユーザーは一度のログインで、許可されたすべてのAWSアカウントやアプリケーションにアクセスできるため、セキュリティが向上し、ユーザーエクスペリエンスが向上します。IAM Identity Centerを使用することで、ユーザーとアクセス権の管理を一元化し、セキュリティの一貫性を保つことができます。

### AWS Directory Service


**AWS Directory Service**は、ディレクトリベースのアイデンティティ管理を提供するサービスです。**Microsoft Active Directory**との統合をサポートし、オンプレミスのディレクトリサービスとシームレスに連携することが可能です。これにより、既存のディレクトリ管理システムをクラウドに拡張し、シングルサインオンやポリシーの一元管理を実現できます。

IAMロールを使用してDirectory Serviceをフェデレートする場合、ユーザーはオンプレミスのディレクトリサービスの資格情報を使用してAWSリソースにアクセスできます。これにより、一貫性のあるセキュリティポリシーを維持しながら、クラウドリソースへのアクセスを簡素化できます。

### Amazon Cognito


**Amazon Cognito**は、ウェブおよびモバイルアプリケーション向けの**エンドユーザー認証**サービスです。サインアップ/サインイン、ユーザープール、**ソーシャルログイン**（Google等）、**SAML/OIDC**によるフェデレーション、**MFA**をサポートします（※通知サービスの **Amazon SNS** とは別物です）。

- **向く例**：スマホアプリやB2C Webのログイン機能を素早く実装
- **向かない例**：AWSコンソールや社内管理者の権限管理 → IAM / IAM Identity Center

## セキュリティ監視と監査
AWSのセキュリティ監視と監査サービスは、システムの安全性を維持し、異常を早期に発見するための重要なツールです。これらのサービスを利用することで、コンプライアンスを確保し、セキュリティインシデントの迅速な対応が可能となります。

### AWS Artifact


**AWS Artifact**は、AWSおよび関連サービスの**コンプライアンス報告書・契約書類**をダウンロードするポータルです。

### AWS Audit Manager


**AWS Audit Manager**は、コンプライアンス監査の証跡収集と評価を**自動化**するサービスです。

### Amazon Detective


**Amazon Detective**は、セキュリティインシデントの調査を支援するサービスです。Detectiveは、**AWS CloudTrailログ、VPC Flow Logs、GuardDutyの検出結果**を分析し、インシデントの原因を特定するのに役立ちます。

### AWS Security Hub


**AWS Security Hub**は、複数のAWSアカウントやリージョンにわたるセキュリティ状態を一元管理するためのサービスです。Security Hubは、セキュリティベストプラクティスに基づいて、セキュリティ状況を評価し、改善のための推奨事項を提供します。

### Amazon GuardDuty


**Amazon GuardDuty**は、脅威検出サービスで、AWSアカウントやワークロードを保護します。GuardDutyは、**AWS CloudTrailイベントログ、VPC Flow Logs、DNSログ**を分析し、異常なアクティビティを検出します。

### Amazon Macie


**Amazon Macie**は、機械学習を利用して、**S3バケット**内の機密データ（例えば、個人情報や金融情報）を自動的に検出し、保護します。Macieを使用することで、データセキュリティを強化し、コンプライアンス要件を満たすことができます。

### Amazon Inspector


**Amazon Inspector**は、EC2やコンテナイメージなどの**脆弱性（CVE）**をスキャンするサービスです。**GuardDuty**が「不審な振る舞い（脅威）」を検知するのに対し、Inspectorは「既知の脆弱性」を見つける用途です。

### AWS Config と AWS CloudTrail（比較）

| サービス | 記録するもの | 主な用途 |
|----------|--------------|----------|
| **CloudTrail** | **誰が**いつ**どのAPI**を呼んだか | 監査・不正調査 |
| **Config** | リソースの**設定状態**とルール違反 | コンプライアンス継続評価 |

## 暗号化とデータ保護
AWSではデータ保護を次の3つに分けて考えます。

- **保管中（at rest）**：KMSでEBS/S3等を暗号化
- **転送中（in transit）**：TLS/ACMでHTTPS化
- **使用中（in use）**：厳格要件では Nitro Enclaves 等（SAAでは概要把握で可）
AWSの暗号化とデータ保護サービスは、データの機密性と完全性を維持するために不可欠です。これらのサービスを使用することで、データの暗号化と安全な管理を実現し、コンプライアンス要件を満たすことができます。

### AWS Key Management Service (AWS KMS)


**AWS Key Management Service (KMS)**は、暗号鍵の作成と管理を簡素化するサービスです。KMSは、データの暗号化と復号化を安全に行うための鍵管理を提供します。これにより、データ保護のための強力なセキュリティが確保されます。

AWS KMSを使用して、保管中のデータを暗号化することが可能です。S3バケットやEBSボリュームに保存されるデータは、KMSの暗号鍵を使用して暗号化されます。これにより、データの機密性が保たれます。

KMSには、暗号化キーに対して詳細なアクセスポリシーを実装する機能があります。これにより、誰がどのような操作を行えるかを細かく制御することが可能です。アクセスポリシーを設定することで、暗号化キーの不正使用を防止できます。

KMSは、暗号化キーのローテーション機能を提供します。定期的にキーを更新することで、セキュリティをさらに強化できます。自動ローテーションを設定することで、手動での管理負担を軽減し、コンプライアンス要件を満たします。

### AWS Certificate Manager (ACM)


**AWS Certificate Manager (ACM)**は、**SSL/TLS証明書**の取得、デプロイ、および管理を簡素化するサービスです。ACMを使用することで、ウェブサイトやアプリケーションの通信を安全に保つための証明書を自動的に更新することができます。

TLSを使用して、データの転送中の暗号化を実現します。これにより、通信途中でのデータ漏洩を防止できます。ACMは、証明書の自動更新機能も提供しており、証明書の有効期限を管理する手間を削減します。

### AWS Secrets Manager


**AWS Secrets Manager**は、**DBパスワードやAPIキー**などの機密文字列を保管し、自動ローテーションできるサービスです。

**Systems Manager Parameter Store**（本章では補足）は設定値の保管にも使えますが、**高度なローテーションや監査が必要な秘密情報は Secrets Manager**、非機密の設定値は Parameter Store、という切り分けが試験でよく問われます。

### AWS CloudHSM


**AWS CloudHSM**は、専用の**ハードウェアセキュリティモジュール（HSM）**上で暗号鍵を管理します。KMSより厳格なコンプライアンス（自社で鍵を完全管理したい場合など）向けです。鍵のバックアップ・マルチAZ配置により可用性を確保します。

## ネットワークセキュリティ
AWSのネットワークセキュリティサービスは、ネットワークレベルでの保護を提供し、不正アクセスや攻撃からシステムを守ります。

**セキュリティグループ（SG） vs ネットワークACL（NACL）**

| | セキュリティグループ | NACL |
|--|---------------------|------|
| 適用単位 | ENI（インスタンス） | サブネット |
| 状態 | ステートフル | ステートレス |
| デフォルト | 全拒否（許可のみ記述） | 許可・拒否ルール |

**IAM Access Analyzer**：リソースが外部（インターネットや他アカウント）に公開されていないかを検出します。

### AWS Firewall Manager


**AWS Firewall Manager**は、AWS WAF、AWS Shield Advanced、VPCセキュリティグループなどのセキュリティポリシーを一元管理するサービスです。Firewall Managerを使用することで、組織全体で統一されたセキュリティポリシーを適用し、管理を簡素化できます。

### AWS Network Firewall


**AWS Network Firewall**は、マネージド型のネットワークファイアウォールサービスです。Network Firewallを使用することで、VPC内のトラフィックを詳細に制御し、侵入防止システム（IPS）やルールベースのフィルタリングを実現できます。

### AWS Shield


**AWS Shield**は、分散型サービス拒否（DDoS）攻撃からアプリケーションを保護するためのサービスです。Shieldには、基本的な保護を提供するAWS Shield Standardと、より高度な保護を提供するAWS Shield Advancedがあります。

### AWS WAF


**AWS WAF**は、ウェブアプリケーションファイアウォールであり、SQLインジェクションやクロスサイトスクリプティング（XSS）などの一般的な攻撃からアプリケーションを保護します。WAFを使用することで、カスタムルールを作成し、特定の攻撃パターンを防御できます。

| サービス | 主に防ぐもの | レイヤ |
|----------|--------------|--------|
| **WAF** | SQLi、XSS などアプリ攻撃 | レイヤ7 |
| **Shield** | DDoS | ネットワーク〜アプリ層 |
| **Network Firewall** | VPC内のIP/ポートフィルタ、IPS | レイヤ3〜7 |

### 試験で押さえる設計の考え方（本章関連）
- **最小権限**：IAMロール・ポリシー・Permission Boundary
- **防御の多層化**：SG + WAF + Shield + KMS

## まとめ
- **IAM / Identity Center / Directory / Cognito**：誰が何にアクセスするか（人・社内・エンドユーザーで使い分け）
- **Config / CloudTrail**：設定評価 vs API監査
- **SG / NACL**：インスタンス vs サブネットのファイアウォール
- **GuardDuty**：脅威検知 / **Inspector**：脆弱性スキャン / **Macie**：S3の機密データ
- **KMS**：暗号鍵 / **ACM**：TLS証明書 / **Secrets Manager**：APIキー等
- **WAF・Shield・Network Firewall**：ネットワークとアプリの防御層

## 復習クイズ　解答・解説

**A1.** **IAMロール**  
EC2/Lambdaにロールを付与し、STSで一時クレデンシャルを取得。→ IAM

**A2.** **IAMユーザー**  
人向けの恒久ID。サービス間はロール。→ IAM

**A3.** **Amazon Cognito**  
エンドユーザー認証。AWSコンソール操作はIAM / IAM Identity Center。→ Amazon Cognito

**A4.** **Amazon Macie**  
S3内の機密データ検出。→ セキュリティ監視と監査

**A5.** **Amazon GuardDuty**（脅威検知） / **Amazon Inspector**（脆弱性スキャン）  
GuardDutyはログ分析による異常検知、InspectorはCVE等。→ セキュリティ監視と監査

**A6.** **AWS KMS（Key Management Service）**  
暗号鍵の作成・管理。ACMはTLS証明書用。→ 暗号化とデータ保護

**A7.** **AWS Secrets Manager**  
DB認証情報のローテーション向け。Parameter Storeは設定値・非機密向けが多い。→ 暗号化とデータ保護

**A8.** **AWS WAF**（SQLi等・レイヤ7） / **AWS Shield**（DDoS）  
Network FirewallはVPC内のトラフィック制御。→ ネットワークセキュリティ

**A9.** **AWS Config**  
設定のコンプライアンス評価。CloudTrailはAPI呼び出しの監査ログ。→ Config と CloudTrail（比較）
