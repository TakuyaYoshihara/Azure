## アイデンティティとアクセス管理

**この章で覚えること**

- **IAM**（人・ロール・ポリシー）／**Cognito**（アプリのエンドユーザー）
- EC2→S3 は **IAMロール**（試験定番）

### AWS Identity and Access Management (IAM)

**ひとこと**：AWSリソースへのアクセスを「誰が・何をできるか」で制御する

**AWS Identity and Access Management (IAM)**

AWSリソースへのアクセスを制御するサービスです。

ユーザー、グループ、ロールに**ポリシー（JSONの権限定義）**を割り当て、**最小権限の原則**（必要な操作だけ許可）で設計します。

#### 基本要素

| 要素 | 役割 | 例 |
|------|------|-----|
| ユーザー | 人やアプリ用の恒久ID | 開発者アカウント |
| グループ | ユーザーをまとめてポリシー付与 | Developers |
| ロール | 一時的な権限。サービスや他アカウント向け | EC2がS3へアクセス |
| ポリシー | 許可/拒否のルール | `s3:GetObject` のみ許可 |

#### 試験の定番

> **試験の定番**
>

> EC2やLambdaからS3へアクセスするときは、アクセスキーを埋め込まず**IAMロール**を使います。
>

> #### ポリシーの種類
>

> **IAMポリシー vs リソースベースポリシー**：IAMポリシーはユーザー/ロールに付与。S3の**バケットポリシー**のようにリソース側に付与するものを**リソースベースポリシー**という。
> 両方が評価される。
>

> **Permission Boundary**：ユーザー/ロールに「最大限これ以上の権限は付与できない」上限を設定する。ルートユーザーは日常運用では使わず、MFA必須とする。
>

> #### MFA・ルートユーザー
>

>

> #### IAMロール（STS・クロスアカウント）
>

> IAMでは、ユーザー、グループ、ロール、ポリシーを組み合わせることで、柔軟な認証モデルを設計できます。**IAMロール**は、一時的なアクセス権を提供するため、特定のタスクを実行する際に非常に有用です。
> ロールは、**AWS Security Token Service（AWS STS）**を使用して一時的なセキュリティ資格情報を取得できます。
>

> IAMロールは、異なるAWSアカウント間でアクセスを共有するために使用されます。**クロスアカウントアクセス**を設定することで、一つのアカウントのリソースに対して他のアカウントのユーザーがアクセスできるようになります。

### AWS IAM Identity Center (AWS Single Sign-On)

**AWS IAM Identity Center（AWS Single Sign-On）**

複数のAWSアカウントやサードパーティアプリケーションへの**シングルサインオン（SSO）**を提供します。

ユーザーは一度のログインで、許可されたすべてのAWSアカウントやアプリケーションにアクセスできるため、セキュリティが向上し、ユーザーエクスペリエンスが向上します。

### AWS Directory Service

**AWS Directory Service**

ディレクトリベースのアイデンティティ管理を提供するサービスです。

**Microsoft Active Directory**との統合をサポートし、オンプレミスのディレクトリサービスとシームレスに連携することが可能です。

### Amazon Cognito

**Amazon Cognito**

ウェブおよびモバイルアプリケーション向けの**エンドユーザー認証**サービスです。

サインアップ/サインイン、ユーザープール、**ソーシャルログイン**（Google等）、**SAML/OIDC**によるフェデレーション、**MFA**をサポートします（※通知サービスの **Amazon SNS** とは別物です）。

- **向く例**：スマホアプリやB2C Webのログイン機能を素早く実装
- **向かない例**：AWSコンソールや社内管理者の権限管理 → IAM / IAM Identity Center

## セキュリティ監視と監査

**この章で覚えること**

- **GuardDuty**＝脅威検知、**Inspector**＝脆弱性スキャン
- **Config**＝設定評価、**CloudTrail**＝API監査

### AWS Artifact

**AWS Artifact**

AWSおよび関連サービスの**コンプライアンス報告書・契約書類**をダウンロードするポータルです。

### AWS Audit Manager

**AWS Audit Manager**

コンプライアンス監査の証跡収集と評価を**自動化**するサービスです。

### Amazon Detective

**Amazon Detective**

セキュリティインシデントの調査を支援するサービスです。

Detectiveは、**AWS CloudTrailログ、VPC Flow Logs、GuardDutyの検出結果**を分析し、インシデントの原因を特定するのに役立ちます。

### AWS Security Hub

**AWS Security Hub**

複数のAWSアカウントやリージョンにわたるセキュリティ状態を一元管理するためのサービスです。

Security Hubは、セキュリティベストプラクティスに基づいて、セキュリティ状況を評価し、改善のための推奨事項を提供します。

### Amazon GuardDuty

**Amazon GuardDuty**

脅威検出サービスで、AWSアカウントやワークロードを保護します。

GuardDutyは、**AWS CloudTrailイベントログ、VPC Flow Logs、DNSログ**を分析し、異常なアクティビティを検出します。

### Amazon Macie

**Amazon Macie**

機械学習を利用して、**S3バケット**内の機密データ（例えば、個人情報や金融情報）を自動的に検出し、保護します。

### Amazon Inspector

**Amazon Inspector**

EC2やコンテナイメージなどの**脆弱性（CVE）**をスキャンするサービスです。

**GuardDuty**が「不審な振る舞い（脅威）」を検知するのに対し、Inspectorは「既知の脆弱性」を見つける用途です。

### AWS Config と AWS CloudTrail（比較）

| サービス | 記録するもの | 主な用途 |
|----------|--------------|----------|
| **CloudTrail** | **誰が**いつ**どのAPI**を呼んだか | 監査・不正調査 |
| **Config** | リソースの**設定状態**とルール違反 | コンプライアンス継続評価 |

## 暗号化とデータ保護

**この章で覚えること**

- **KMS**
- **ACM**
- **Secrets Manager**
- **CloudHSM**

AWSではデータ保護を次の3つに分けて考えます。

- **保管中（at rest）**：KMSでEBS/S3等を暗号化
- **転送中（in transit）**：TLS/ACMでHTTPS化
- **使用中（in use）**：厳格要件では Nitro Enclaves 等（SAAでは概要把握で可）

### AWS Key Management Service (AWS KMS)

**AWS Key Management Service (KMS)**

- 暗号鍵の作成・管理（EBS/S3 等の at rest 暗号化）
- 鍵のアクセスポリシー、自動ローテーション

### AWS Certificate Manager (ACM)

**AWS Certificate Manager (ACM)**

**SSL/TLS証明書**の取得、デプロイ、および管理を簡素化するサービスです。

### AWS Secrets Manager

**AWS Secrets Manager**

**DBパスワードやAPIキー**などの機密文字列を保管し、自動ローテーションできるサービスです。

- **Secrets Manager**：DBパスワード・APIキー等の秘密情報、自動ローテーション
- **Parameter Store**：非機密の設定値（試験では Secrets と役割分担）

### AWS CloudHSM

**AWS CloudHSM**

専用の**ハードウェアセキュリティモジュール（HSM）**上で暗号鍵を管理します。

KMSより厳格なコンプライアンス（自社で鍵を完全管理したい場合など）向けです。鍵のバックアップ・マルチAZ配置により可用性を確保します。

## ネットワークセキュリティ

**この章で覚えること**

- **WAF**（Web）
- **Shield**（DDoS）
- **Network Firewall**（VPC）

**セキュリティグループ（SG） vs ネットワークACL（NACL）**

| | セキュリティグループ | NACL |
|--|---------------------|------|
| 適用単位 | ENI（インスタンス） | サブネット |
| 状態 | ステートフル | ステートレス |
| デフォルト | 全拒否（許可のみ記述） | 許可・拒否ルール |

**IAM Access Analyzer**：リソースが外部（インターネットや他アカウント）に公開されていないかを検出します。

### AWS Firewall Manager

**AWS Firewall Manager**

AWS WAF、AWS Shield Advanced、VPCセキュリティグループなどのセキュリティポリシーを一元管理するサービスです。

### AWS Network Firewall

**AWS Network Firewall**

マネージド型のネットワークファイアウォールサービスです。

### AWS Shield

**AWS Shield**

分散型サービス拒否（DDoS）攻撃からアプリケーションを保護するためのサービスです。

Shieldには、基本的な保護を提供するAWS Shield Standardと、より高度な保護を提供するAWS Shield Advancedがあります。

### AWS WAF

**AWS WAF**

ウェブアプリケーションファイアウォールであり、SQLインジェクションやクロスサイトスクリプティング（XSS）などの一般的な攻撃からアプリケーションを保護します。

| サービス | 主に防ぐもの | レイヤ |
|----------|--------------|--------|
| **WAF** | SQLi、XSS などアプリ攻撃 | レイヤ7 |
| **Shield** | DDoS | ネットワーク〜アプリ層 |
| **Network Firewall** | VPC内のIP/ポートフィルタ、IPS | レイヤ3〜7 |

> ### 試験で押さえる設計の考え方（本章関連）
>
>

> - **最小権限**：IAMロール・ポリシー・Permission Boundary
> - **防御の多層化**：SG + WAF + Shield + KMS

## まとめ

- **IAM / Identity Center / Directory / Cognito**：誰が何にアクセスするか（人・社内・エンドユーザーで使い分け）
- **Config / CloudTrail**：設定評価 vs API監査
- **SG / NACL**：インスタンス vs サブネットのファイアウォール
- **GuardDuty**：脅威検知 / **Inspector**：脆弱性スキャン / **Macie**：S3の機密データ
- **KMS**：暗号鍵 / **ACM**：TLS証明書 / **Secrets Manager**：APIキー等
- **WAF・Shield・Network Firewall**：ネットワークとアプリの防御層

## 復習クイズ　解答・解説

**A1.** **IAMロール**  
EC2/Lambdaにロールを付与し、STSで一時クレデンシャルを取得。→ IAM

**A2.** **IAMユーザー**  
人向けの恒久ID。サービス間はロール。→ IAM

**A3.** **Amazon Cognito**  
エンドユーザー認証。AWSコンソール操作はIAM / IAM Identity Center。→ Amazon Cognito

**A4.** **Amazon Macie**  
S3内の機密データ検出。→ セキュリティ監視と監査

**A5.** **Amazon GuardDuty**（脅威検知） / **Amazon Inspector**（脆弱性スキャン）  
GuardDutyはログ分析による異常検知、InspectorはCVE等。→ セキュリティ監視と監査

**A6.** **AWS KMS（Key Management Service）**  
暗号鍵の作成・管理。ACMはTLS証明書用。→ 暗号化とデータ保護

**A7.** **AWS Secrets Manager**  
DB認証情報のローテーション向け。Parameter Storeは設定値・非機密向けが多い。→ 暗号化とデータ保護

**A8.** **AWS WAF**（SQLi等・レイヤ7） / **AWS Shield**（DDoS）  
Network FirewallはVPC内のトラフィック制御。→ ネットワークセキュリティ

**A9.** **AWS Config**  
設定のコンプライアンス評価。CloudTrailはAPI監査ログ。→ セキュリティ監視と監査
