## リレーショナルデータベース

**この章で覚えること**

- **RDS** / **Aurora**・**Multi-AZ** vs **リードレプリカ**

### Amazon RDS

**Amazon RDS（Relational Database Service）**

クラウド上でリレーショナルデータベースを簡単にセットアップ、運用、およびスケーリングするためのサービスです。

MySQL、PostgreSQL、MariaDB、Oracle、SQL Serverなどをサポートします。**自動バックアップ**、**手動スナップショット**、メンテナンスウィンドウ、パッチ適用をAWSが代行します。

### Amazon Aurora

**Amazon Aurora**

MySQLおよびPostgreSQL互換の高性能RDBです。

ストレージは自動拡張し、AZ内で6コピー保持されます。**Aurora Global Database**は、プライマリリージョンから最大5つのセカンダリリージョンへレプリケーションし、DRと低RPOを実現します（試験頻出）。

### Amazon Aurora Serverless

**Amazon Aurora Serverless**

負荷に応じてキャパシティを自動調整するオプションです（現在は **Aurora Serverless v2** が一般的）。

断続的なアクセスや予測しづらい負荷向けです。

> **Multi-AZ とリードレプリカ（試験頻出）**
>
>

> | 機能 | 目的 |
> |------|------|
> | **Multi-AZ** | 障害時の自動フェイルオーバー（可用性） |
> | **リードレプリカ** | 読み取り負荷の分散（最大5台など） |

- **向く例（RDS/Aurora）**：トランザクション、既存RDB移行、複雑なSQL
- **向かない例**：秒間数百万のキー・バリューアクセス → DynamoDB

## NoSQLデータベース

**この章で覚えること**

- **DynamoDB**
- **DAX**
- DocumentDB / Keyspaces

### Amazon DynamoDB

**Amazon DynamoDB**

フルマネージドなNoSQL（キー・バリュー）データベースで、ミリ秒台の応答が特徴です。

**オンデマンド**（使った分）と**プロビジョンド**（RCU/WCUを指定）のキャパシティモードがあります。DynamoDB Streamsで変更イベントをLambda等に連携できます。

- **DynamoDB Accelerator（DAX）**：読み取り専用キャッシュで読取レイテンシをさらに短縮
- **グローバルテーブル**：マルチリージョンでレプリケーション（マルチマスター）
- **DocumentDB** との違い：JSONドキュメント・MongoDB互換が必要なら DocumentDB、キー・バリューでスケールなら DynamoDB

### Amazon DocumentDB (MongoDB互換)

**Amazon DocumentDB（MongoDB互換）**

スケーラブルで高可用性のドキュメントデータベースサービスです。

DocumentDBは、MongoDBのAPIと互換性があり、MongoDBデータベースからの移行を容易にします。フルマネージドなサービスであり、バックアップやスケーリング、パッチ適用などの管理タスクを自動化します。

### Amazon Keyspaces (Apache Cassandra向け)

**Amazon Keyspaces（Apache Cassandra向け）**

スケーラブルで高可用性のNoSQLデータベースサービスで、Apache Cassandra互換のAPIを提供します。

Keyspacesは、Cassandraクラスターの管理を簡素化し、インフラストラクチャの運用を軽減します。

## インメモリキャッシュ

**この章で覚えること**

- **ElastiCache**（Redis vs Memcached）

### Amazon ElastiCache

**Amazon ElastiCache**

RedisおよびMemcachedのインメモリキャッシュです。

| エンジン | 特徴 | 向く例 |
|----------|------|--------|
| **Redis** | 永続化、レプリケーション、データ構造が豊富 | セッション、ランキング、Pub/Sub |
| **Memcached** | シンプルなキー・バリューキャッシュ | 読み取りキャッシュのみ |

## グラフデータベース

**この章で覚えること**

- **Neptune**（関係性データ）

### Amazon Neptune

**Amazon Neptune**

フルマネージドなグラフデータベースサービスで、グラフモデルに基づいたデータの格納とクエリをサポートします。

Neptuneは、Property GraphとRDF（Resource Description Framework）をサポートし、複雑なグラフクエリを効率的に実行します。高可用性と耐障害性を備えており、大規模なグラフデータセットに対しても優れたパフォーマンスを発揮します。

## 台帳・監査向けデータベース

**この章で覚えること**

- **QLDB**（改ざん不能な台帳）

### Amazon Quantum Ledger Database (Amazon QLDB)

**Amazon QLDB**

**改ざん不能な台帳（レジャー）**として全変更履歴を記録するデータベースです。

金融取引や監査証跡など、履歴の完全性が重要な用途向けです（一般的なトランザクションDBの代替ではありません）。

## データウェアハウス

**この章で覚えること**

- **Redshift**（OLAP）
- **Athena** / **Glue**（S3 分析）

### Amazon Redshift

**Amazon Redshift**

OLAP（分析）向けのデータウェアハウスです。

RDSなどの**OLTP（取引処理）**とは用途が異なります。Redshift SpectrumでS3上のデータもSQLから参照できます。

- **向く例**：BI、集計レポート、ログ分析
- **向かない例**：1件ずつの更新が多い業務システム → RDS/Aurora

> **S3上のデータ分析（試験補足）**
>
>

> - **Amazon Athena**：S3のデータに対してサーバーレスでSQLクエリ（ログ分析等）
> - **AWS Glue**：ETLジョブとデータカタログ（メタデータ管理）

## コスト最適化とキャパシティプランニング

**この章で覚えること**

- **RDS Proxy**
- キャパシティ設計

DB設計では、コストの見える化とキャパシティの見積もりを押さえます。

### コスト管理サービス

DBコストを部門・プロジェクト単位で追いやすくする仕組みです。

- **コスト配分タグ**：リソースにタグを付け、コストレポートで按分
- **マルチアカウント請求**（Organizations）：複数アカウントの請求を統合

### コスト管理ツール

コストの「見える化」と「予算アラート」に使う代表的なツールです（詳細はガバナンス章）。

- **Cost Explorer**：コスト・使用量をグラフ化し傾向を分析
- **AWS Budgets**：予算・しきい値を超えたらアラート
- **AWS Cost and Usage Report**：詳細なコスト・利用レポート（分析・請求照合用）

### キャッシュ戦略

読み取りをキャッシュして DB 負荷とコストを下げます（詳細は本章「インメモリキャッシュ」）。

- **Amazon ElastiCache**：Redis / Memcached のマネージドキャッシュ

### データ保持ポリシー

不要データを減らし、ストレージコストを抑える考え方です。

- **データのライフサイクル**：古いデータをアーカイブまたは削除
- **（関連）S3 ライフサイクル**：オブジェクトの保管クラス移行・削除（ストレージ章）

### キャパシティプランニング

サービスごとにキャパシティの単位が異なります。

| サービス | 単位の例 |
|----------|----------|
| **DynamoDB** | RCU / WCU、またはオンデマンド |
| **Aurora Serverless** | ACU |

### データベース接続とプロキシ

接続数が増える構成では、プールで枯渇を防ぎます。

- **Amazon RDS Proxy**：DB接続をプールし、スケール時の接続枯渇を緩和

### データベースエンジンの適切なユースケース

RDS で選べるエンジンの特性の違い（試験では概要把握で十分なことが多い）。

| エンジン | 向く例 |
|----------|--------|
| **MySQL** | 読み取り性能、一般的な Web |
| **PostgreSQL** | 複雑な SQL、整合性・拡張を重視するアプリ |

> ### 試験で押さえる設計の考え方（本章関連）
>
>

> - **可用性**：Multi-AZ（RDS）、Aurora Global Database（DR）
> - **性能**：リードレプリカ、ElastiCache、DAX
> - **分析**：Redshift（OLAP）、Athena（S3 SQL）

## まとめ

- **RDS / Aurora**：SQL・トランザクション。Multi-AZ＝可用性、リードレプリカ＝読取分散
- **Aurora Global Database / DAX**：DR・読取高速化
- **DynamoDB**：スケールするNoSQL。オンデマンド vs プロビジョンド
- **Athena / Glue**：S3分析・ETL（試験補足）
- **ElastiCache**：Redis vs Memcached でキャッシュ戦略
- **Neptune**：グラフ（関係性）データ
- **QLDB**：改ざん不能な台帳・監査
- **Redshift**：分析（OLAP）。RDS（OLTP）と用途を混同しない
- **RDS Proxy**：接続プールでスケール時の接続枯渇を緩和

## 復習クイズ　解答・解説

**A1.** **Multi-AZ**  
スタンバイへの自動フェイルオーバー（可用性）。読取分散ではない。→ リレーショナルデータベース

**A2.** **リードレプリカ（Read Replica）**  
読み取り専用の複製で読取負荷を分散。→ リレーショナルデータベース / コスト最適化節  
※よくある誤答：Multi-AZ（可用性が目的）

**A3.** **Amazon DynamoDB**  
スケーラブルなNoSQL。RDSはSQL・トランザクション向け。→ NoSQLデータベース

**A4.** **Amazon RDS**（または **Amazon Aurora**）  
マネージドRDB。→ リレーショナルデータベース

**A5.** **Redis**  
永続化・レプリケーション・豊富なデータ構造。Memcachedはシンプルなキャッシュのみ。→ インメモリキャッシュ

**A6.** **Amazon Redshift**  
OLAP・分析。RDSはOLTP（取引処理）。→ データウェアハウス

**A7.** **Amazon QLDB（Quantum Ledger Database）**  
改ざん不能な変更履歴・台帳。一般的な分散DBの代替ではない。→ 分散データベース

**A8.** **Amazon RDS Proxy**  
接続プールでスケール時の接続枯渇を緩和。→ コスト最適化とキャパシティプランニング

**A9.** **DynamoDB Accelerator（DAX）**  
DynamoDB専用の読取キャッシュ。→ Amazon DynamoDB（DAXの説明）
